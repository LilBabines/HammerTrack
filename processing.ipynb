```python
import pandas as pd
import os
```

```python

for f in os.listdir("projects/hammer/exports/2021_114/per_track_csv"):

    df1 = pd.read_csv("projects/hammer/exports/2021_114/per_track_csv/"+f)
    df2 = pd.read_csv("projects/hammer/exports/2021_114/keypoints/"+f)

    df2 = df2[["frame","head_x","head_y","com_x","com_y","art1_x","art1_y","art2_x","art2_y","tail_x","tail_y"]]
    m = pd.merge(df1, df2, left_on='frame', right_on='frame', how='left',suffixes=("",""),)

    m = m.loc[:, ~m.columns.str.contains('^Unnamed')]

    m.to_csv("projects/hammer/exports/2021_114/with_keypoints/"+f, index=False)

    

```
